#`matplotllib` For Beginners 

For us in the academic field (and also in our future careers) it is important to analyze our data through visualization. Python has its essential plotting library: `matplotlib`, and in this module we discuss its functions and properties.

#Importing `matplotlib`

To start, we call out `matplotlib` library using the `import` function. A common concise assignment to it is `mpl`.

```python
import matplotlib as mpl
```

And then we call a sublibrary of it: `pyplot`, which gives ourselves extra function commands that manipulate our plots in many ways and to be similar to MATLAB (another computational language). A common concise assignment to it is `plt`

```python
import matplotlib.pyplot as plt
```

# Making Your First Plot

Test out this code below, or make it your own. To make your own (2D) plot, you have to assign a set of values per variable (x and y), then use the `plt.plot(#xvalues, #yvalues)` command to input your x and y values into the plot. And in order to show it, we use the `plt.show()` function.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) #why are these discrete number set when the resulting graph is a continuous function? 
y = x
#the commands to make plot

plt.plot(x, y)
plt.plot(x, np.cos(x)) #notice that since we have two plt.plot() commands, the code automatically assigns different colors per graph
plt.show()
```

Congratulations! You have done the first step to generate a simple plot. `matplotlib` has a lot ways to do plots such as curve-fitting, population histogram, and 3D plots (All of these plots can be saved as images). We start first with curve fitting.

#Scatter Plots: Making Your First Curve Fit

Let us make the essentials of any experiment which is the scatter plot with error bars and fitting curve. It is very common in experiments to adjust the axes, insert axes titles as well as general titles, input legends, and to make a trendline with an equation.  

Let us assign data points and plot them with a predetermined function.

We can simply command `plt.scatter(xpoints, ypoints)` to discretize our previously made function into a scatter plot.



```python
x = np.linspace(0, 10, 10) #We use a fewer number of data points to clearly see the scatter
y = np.cos(x)

plt.scatter(x, y) #more on this function later
plt.show()
```

But we can also add an argument in a `plt.plot()` function to make it discretized.

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid') # this is used to make a grid in the plot

x = np.linspace(0, 10, 20)
y = x

plt.plot(x, y, "o", color = "red") #"o" makes the graph into a scatter plot
plt.show()
#next is we put the data points individually, and eventually compute a fitting line after (without the manual Jacobian computation)

```

We have other options of the shapes that we can assign to each data point for the purpose of differentiating between the variables you are studying. For example we can put `'.', ',', 'x', '+', 'v', '^', '<', '>', 's',` or `'d'` on our scatter plot (try it!).

We have a sample code showing these characters 

```python
x = np.linspace(0, 10, 5)
y = x

for character in ['.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
  y = y + 2
  plt.plot(x, y, character) #, label = ': "{0}"'.format(character) for label
#plt.legend(numpoints=1) #we add legend here
plt.show()
```

Notice here below that when we add a hypen to each character, the data points are traced. 

```python
x = np.linspace(0, 10, 5)
y = x

for character in ['-.', '-,', '-x', '-+', '-v', '-^', '-<', '->', '-s', '-d']:
  y = y + 2
  plt.plot(x, y, character) #, label = ': "{0}"'.format(character) for label
#plt.legend(numpoints=1) #we add legend here
plt.show()
```

We can also add arguments to `plt.plot()` that edit the characteristics of our plot manually instead of just the markers

```python
x = np.linspace(0, 10, 20) #We use a fewer number of data points to clearly see the scatter
y = np.cos(x)

plt.plot(x, y, '-p', color='magenta',
         markersize=15, linewidth=4,
         markerfacecolor='chartreuse',
         markeredgecolor='magenta',
         markeredgewidth=2) #more on this function later
plt.show()
```

We assign a color string to the color variable, then the other arguments: markersize, linewidth, markerfacecolor, markeredgecolor, and markeredgewidth, are self-explanatory. More details on which arguments you can put and which marker symbols you can assign are detailed in the matplotlib library info.

Previously, we have used `plt.scatter()` by which without a marker argument, it already creates a scatter plot. Here we explore more of its advantage which is that we can assign a different property for EACH data point. 

```python
plt.scatter(x, y, marker='d') #Here we assign the 'd' string to the marker variable for plt.scatter() function
```

Here we can manipulate each data point by putting a sample array of color and size values (in random from 0 to 1).

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 500 * rng.rand(100)

plt.scatter(x,y, c = colors, s = sizes, alpha = 0.3, cmap = 'hot')
#notice the new arguments for color and sizes are stored in variables c and s respectively, alpha gives transparency value,
#and cmap for color set (if there is none we have a grayscale value)
plt.colorbar();
```

Matplotlib has a lot of other graph types for our convenience. And unlike other softwares like Excel/Gsheets we can take in different data formats for graphing including arrays and dataframes (in pandas). We see here a sample plot imported from a sample data

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.4,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
plt.colorbar();
```

Now we edit the essential parts of the plot. First are the axis labels where we simply call out for x and y axes respectively: `plt.xlabel("name string")` and `plt.ylabel("name string")`. For the title we use `plt.title("Position Graph of Car over time")`

For axes scaling we use `plt.xlim(lower bound, higher bound)` and `plt.ylim(lower bound, higher bound)`.  Let's edit our previous scatter plot but with the following features 

```python
plt.style.use('seaborn-whitegrid') 

x = np.linspace(0, 10, 20)
y = x

plt.scatter(x, y, marker = "o", color = "red") 
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.ylim(4, 10)
```

We add errors by calling `plt.errorbar(x, y, yerr=dy)`. For errors along the y axis assign the error value to argument `yerr` and `xerr` for x axis error.

```python
plt.style.use('seaborn-whitegrid') 

x = np.linspace(0, 10, 20)
y = x
dy = 0.8


plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.title("Position Graph of Car over time")
plt.errorbar(x, y, marker = 'o', yerr=dy, fmt='.k') #more details on manipulating errorbar design on the reference
```

For editing two subplots with a legend, we can use this way of plotting:

```python
x = np.linspace(0, 10, 20)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
leg = ax.legend(loc='upper left', frameon=True); 
#will automatically put the legend on the top right corner. Now try adding an argument inside the parentheses loc='upper left', frameon=True
```

Through this way we can rescale the axes (not changing the limits) in many ways such as the following

```python
x = np.linspace(0, 10, 20)
fig, ax = plt.subplots()
ax.plot(x, np.exp(x), '-b', label='Sine')
ax.plot(x, 2**(x), '--r', label='Cosine')
ax.set_yscale('log') #sets the yscale into log scale
ax.set_ylabel('in log scale')
ax.axis('equal')
leg = ax.legend(loc='upper left', frameon=True); 
#will automatically put the legend on the top right corner. Now try adding an argument inside the parentheses loc='upper left', frameon=True
```
