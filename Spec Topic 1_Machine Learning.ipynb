### What is Machine Learning?

In typical programming, we explicitly tell the computer which algorithm to solve which problem. In machine learning, all we tell the computer is how it can teach itself how to reach a solution. The computer, using data that you give it, identifies patterns and makes decisions that makes a model that most accurately solves the problem you gave.

### Import the Libraries

Scikit-learn, or `sklearn`, is a free machine-learning library for Python. It is simple, efficient, and easy-to-read in various contexts. 

```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

### Load in the Datasets

In the `sklearn` module, there are various datasets that you can use in order to try out the different features of the library. They are built-in functions in the library that you can call on whenever. One of the most commonly used libraries is the Iris Dataset.

The Iris Dataset is the best known database in pattern recognition and machine learning. The dataset contains 3 types of Iris Plant (classes) of 50 instances each, and we are given each sample's measurements: the sepal length, sepal width, petal length, petal width, and the type of Iris Plant.

Our goal is to make a model where the program can determine the kind of iris plant based from the sepal and petal lengths and widths.

```python
iris = datasets.load_iris()
```

The dataset `iris` we have just defined is actually a pair of variables. There is the `data` which contain the features we'll train the model on: the sepal and petal lengths and widths; and the `target`, the feature that we want to be able to predict, in this case the type of plant. We define these separately.

```python
X = iris.data
y = iris.target
```

### The Machine Learning Model

There are many different models you can use in order to solve a machine learning problem. The skill of knowing which one to use is something a machine learning scientist takes years to train. There is no one size fits all model, and which model fits which problem can be very debatable.

In this case, we choose the Decision Tree Classifier. As you will see later, it looks like an upside-down tree, with decisions at each branching point. When the condition is satisfied (True), go left; otherwise, go right. This simple model, when used correctly, can be a very powerful machine learning tool. Here, we define our Decision Tree Classifier.

```python
decision_tree = DecisionTreeClassifier()
```

### Fitting the Machine Learning Model

Here is the single line that will do most of the work in the machine learning program, and the work that the `scikit-learn` library automates and makes easy for us. The `fit` function tells the computer, here is some data `X` that is supposed to correctly predict `y`; find a pattern using this model `clf`. 

```python
model = decision_tree.fit(X, y)
```

### Visualizing the Model we Made

Here, I visualize the model that we've created. Note that I did not give the computer or program anything besides two things: the data, and the kind of model I wanted it to use. The computer made this model through a tedious process of trial and error, giving us a model that is as accurate as possible. At each branching point, the "condition" is the first line and if that condition is satisfied, go left; otherwise, go right. This gives us a very easy to follow process of how to classify Iris plants.

The other lines are metrics describing how the model performs.
1. `gini` is a metric that quantifies the "purity" of a node; if it is zero that means the node is pure, and otherwise the node is made up of different 
2. `samples` is the number of remaining data points remain to be classified in each node
3. `class` in a node refers to the class that occurs most in that node

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(
    decision_tree, 
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True
)
```

### How Good is our Model?

At the end, we want to know if the model we made is accurate at all. We wanna see that if we predict all the datapoints using the model we made would they all give the correct answer? The `scikit-learn` library has built-in functions just for that too.

```python
from sklearn.metrics import accuracy_score
```

We want to use the `accuracy_score` function between the predictions we've made and the actual answers. Thus, we define `predictions` and plug that and `y` into the function. We see that we have a score of 100, and we have perfect accuracy!

```python
predictions = model.predict(X)

score = accuracy_score(predictions, y)*100
print(score)
```
